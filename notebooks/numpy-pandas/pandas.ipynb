```python
# 各行が条件にマッチするか判定
titanic_df['Age'] > 20
```

```python
# 条件にマッチする行を抽出
titanic_df[titanic_df['Age'] > 20]
```

```python
# 複数条件にマッチする行を抽出
titanic_df[(titanic_df['Age'] > 20) & (titanic_df['Sex'] == 'female')]
```

```python
# query関数を使用
titanic_df.query('(Age > 20) & (Sex == "female")')
```
